```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#Q.Write a python program to display IMDB’s Top rated 100 Indian movies’ data https://www.imdb.com/list/ls056092300/ (i.e. name, rating, year ofrelease) and make data frame
page1 = requests.get('https://www.imdb.com/list/ls055592025/')
imdb= BeautifulSoup(page1.content,"html.parser")
```

```python
#scraping names of the movies

names =[]

for i in imdb.find_all('h3',class_="lister-item-header"):
    names.append(i.get_text().strip("\n").replace("\n","").split("(")[0])
```

```python
#scraping year of the movies

year =[]

for i in imdb.find_all('span',class_="lister-item-year text-muted unbold"):
    year.append(i.text)
    
```

```python
#scraping rating of the movies

ratings =[]

for i in imdb.find_all('div',class_="ipl-rating-star small"):
    ratings.append(i.text.replace("\n",""))
    
```

```python
#Making DataFrame

df= pd.DataFrame({'Movie_Names':names,'Year':year,'Rating':ratings})
df
```

```python
#Q. Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
#a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating
page4= requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
men = BeautifulSoup(page4.content,'html.parser')
```

```python
teams=[]
for i in men.find_all("span",class_="u-hide-phablet"):
    teams.append(i.get_text())
```

```python
matches=[]
matches.append((men.find("td",class_="rankings-block__banner--matches")).text)
x=1
for i in men.find_all("td",class_="table-body__cell u-center-text"):
    if x%2==1:
        matches.append(i.text)
    x=x+1
```

```python
points=[]
points.append((men.find("td",class_="rankings-block__banner--points")).text)
x-=1

for i in men.find_all("td",class_="table-body__cell u-center-text"):
    if x%2==0:
        points.append(i.text)
    x=x+1
```

```python
ratings=[]

for i in men.find_all("td",class_="rankings-block__banner--rating u-text-right"):
    ratings.append(i.get_text().replace("\n","").replace(" ",""))
    
for i in men.find_all("td",class_="table-body__cell u-text-right rating"):
    ratings.append(i.get_text())
```

```python
#Making DataFrame

dfm= pd.DataFrame({'Teams':teams,'Matches':matches,'Points':points,'Ratings':ratings})
dfm[0:10]
```

```python
#Q. Please visit https://www.cnbc.com/world/?region=world and scrapa) headingsb) date c) News link
import pandas as pd
from bs4 import BeautifulSoup
import requests
page8 = requests.get('https://www.cnbc.com/world/?region=world')
news = BeautifulSoup(page8.content,'html.parser')
```

```python
title=[]
for i in news.find_all("a",class_="LatestNews-headline"):
    title.append(i.get_text())
    
time=[]
for i in news.find_all("time",class_="LatestNews-timestamp"):
    time.append(i.get_text())
    
url=[]
for links in news.find_all("a",class_="LatestNews-headline"):
    links=links.get('href')
```

```python
df= pd.DataFrame({'Headline':title,'Time':time,'News link':links})
df
```
